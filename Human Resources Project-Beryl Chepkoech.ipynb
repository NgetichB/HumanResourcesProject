```python
#PCA Dimensionality Reduction
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#31.Standardize numeric features before applying PCA
numeric_cols = df.select_dtypes(include=['number']).columns
scaler = StandardScaler()
X = df.select_dtypes(include=['number']).dropna()
X_scaled = scaler.fit_transform(X)
```

```python
#32.Perform PCA on the dataset and explain the first two components- scikit-learn PCA
scaled_data = scaler.fit_transform(df[numeric_cols].dropna())
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(
    data=pca_components, 
    columns=["PC1", "PC2"]
)
print(pca_df.head())
```

```python
#33.Plot the explained variance ratio of the first two principal components
print(f"Explained variance ratio of the first two principal components: {pca.explained_variance_ratio_}")
```

```python
#34.Reduce the dataset to two dimensions with PCA and plot employees- seaborn scatterplot
sns.scatterplot(x=pca_df['PC1'], y=pca_df['PC2'])
plt.title('PCA of Employee Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
#Which Variables contribute most to the first principal component
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
# Create a DataFrame of loadings
loadings = pd.DataFrame(
    pca.components_.T,  # transpose so variables are rows
    columns=[f"PC{i+1}" for i in range(pca.n_components_)],
    index=X.columns
)
print("Loadings for PC1:")
print(loadings["PC1"].sort_values(ascending=False))
```

```python
#36.Apply PCA to engagemnt + satisfaction +absences, condense into 1 dimension
sns.scatterplot(x=pca_df['PC1'], y=pca_df['PC2'])
plt.title('PCA of Employee Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
#38.Comparison of clustering before and after PCA using k means on numeric data-scikit-learn KMeans(BEFORE PCA)
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('K-Means Clustering Before PCA')
```

```python
#AFTER PCA
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('K-Means Clustering After PCA')
```

```python
#39.Plot loadings of salary, abscences, engagement on the first two PCAs-PCA Biplot
biplot_data = loadings.loc[['Salary', 'Absences', 'EngagementSurvey'], ['PC1', 'PC2']]
plt.figure(figsize=(10, 8))
# Provide arrays of zeros for the starting points
plt.quiver(
    [0]*len(biplot_data), [0]*len(biplot_data),
    biplot_data['PC1'], biplot_data['PC2'],
    angles='xy', scale_units='xy', scale=1, color='r'
)
for i, txt in enumerate(biplot_data.index):
    plt.text(biplot_data['PC1'][i], biplot_data['PC2'][i], txt)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()
```

```python
#Discuss when PCA is appropriate in HR analytics--analytics Vidhya-use cases

#PCA is an unsupervised algorithm that performs dimensionality reduction by transforming a dataset with many variables into a smaller set with fresh variables called principal components
#PCA is useful in HR Analytics whenever there is high dimensional HR data with many correlated variables, hence the data is reduced to afew uncorrelated components. 
#PCA in HR is useful in the following ways;
#1.Correlated variables-PCA transforms variables such as experience years, job level and salary band into independent components to avoid multicollinearity. 
#2.Dimensionality Reduction-HR datasets with numerous features like scores, engagements, training hours, job satisfation and performance; PCA cancels or reduces noise while maintaining the core data/information
#3.Data Visualization and Clustering-PCA allows you to plot data into 2d/3d plots hence makes it easier to detect clusters and identify patterns
#4.PCA helps reduce redundancy and speeds up computation in data Processing-
#The use cases in HR analytics are: Performance evaluation, employee attrition and engagement surveys, compensation benchmarking and recruiment analytics
```
